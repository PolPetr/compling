Допустим, что x - предложение source language, а y - предложение target language. В статистическом машинном переводе происходит анализ результата, то есть через y восстанавливается x, а именно правила, благодаря которым x преобразился в y. Для успешной работы SMT важны три модели: языковая, модель перевода и декодер.
1. В SMT для начала строится N-граммная языковая модель, то есть устанавливается вероятностное распределение соответствия y x. Ищутся возможные x при получившемся y.
2. Представляем, что y является просто искажением x. Задача состоит в том, чтобы с помощью модели перевода найти вероятность соответствия x y.
3. С помощью декодера выбирается грамматически и лексически правдоподобный вариант x по принципу жадного алгоритма: первый значит лучший.
