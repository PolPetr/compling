Система обрабатывает пару языков русский-французский. Будет использован трансферный перевод, поэтому в модель будут внесены некоторые данные о структуре языка

1. Принимаем на вход исходный текст
2. Токенизация текста по пробелу и его приведение к нижнему регистру
3. Частеречная разметка
4. Морфологическая разметка
5. Лексический трансфер
6. Проверка лексической сочетаемости
7. Создать список французских прилагательных, которые стоят перед существительным. Сверяться по этому списку. При переводе все остальные прилагательные, которые в него не входят, ставить после существительного, как полагается во французском.
8. Внести информацию о соотношении видовременных форм: пр. вр. несов. в. в русском = имперфект во французском; пр. вр. сов. в русском = прошедшее завершенное во французском.
9. Внести информацию о роде во множественном числе: если при переводе на французский известно, что nous и vous -- это лица только ж.р., тогда нужно согласовать по роду глагол (добавить e).
10. Можно дополнить про слово *который*: русский номинатив = французское qui, русский аккузатив = французское que, остальные русские косвенные падежи, кроме аккузатива = dont
