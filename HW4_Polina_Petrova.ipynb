Кандидат  -- это переведенное машиной предложение (машинный перевод). В свою очередь, референс -- это перевод, выполненный человеком, и то, с чем сравнивается кандидат для оценки качества машинного перевода.
BLEU-Score подойдет для подсчета результатов работы системы на целом корпусе.
Оценка может представлять из себя любое число от 0 до 1 (либо от 1 до 100), где 0 -- самый низкий скор, перевод кандидата далек от референса, а 1 -- самый высокий, то есть перевод кандидата максимально близок к референсу.

Результаты в ячейках позволяют утверждать, что машинный перевод не вполне качественный, поскольку BLEU-score составляет 0,374. У автоматической системы возникли трудности с коллокациями (*usual hockey*) и пословицами (*You reap what you sow*), которые переводились буквально (+ буквальный перевод *это* как *is* вместо более естественного для английского языка *has to do with*). Исключались повторы слов (*ice* вместо *under-ice*), хотя было принципиально оставить *under*. Поскольку слова переводятся последовательно, порядок слов при машинном переводе в целевом языке остается таким же, как в языке-источнике. По этой причине с *also* в кандидате начинается предложение. При машинном переводе выбирается скорее нейтральная и более частотная лексика, поэтому мы видим в кандидате *small* вместо *scarce*.
